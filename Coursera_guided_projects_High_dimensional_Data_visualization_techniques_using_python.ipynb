```python
import pandas as pd
from sklearn import preprocessing
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import plotly.express as px

from sklearn.decomposition import PCA
```

```python
data=pd.read_csv('Dataset.csv')
```

```python
data
```

```python
data.info()
```

```python
min_max_scaler=preprocessing.MinMaxScaler()
x=data.drop("Country or region",axis=1).values
x_scaled=min_max_scaler.fit_transform(x)


data[data.columns[1:]]=x_scaled
data
```

```python
data.describe()#normalization has been achieved
```

```python
#clustering is the task of grouping a set of objects in such a way that objects in the same group are
#more similar to each other than those in other groups
```

```python
#a country is an object
#we want to group objects with similarity in their features
```

```python
clustering_data=data.drop('Country or region',axis=1)
kmeans=KMeans(n_clusters=3,max_iter=300).fit(clustering_data)

data['class']=kmeans.labels_.astype(str)
```

```python
data
```

```python
data['class'].value_counts()
```

```python
dims=data.drop(['Country or region','class'],axis=1).columns
fig=px.scatter_matrix(data,dimensions=dims,color=data['class'],hover_name=data['Country or region'])

fig.update_traces(marker=dict(size=4),diagonal_visible=False,showupperhalf=False)#to make the dots smaller
fig.update_layout(width=800,height=800,title='Happiness index SPLOM',font_size=7)
fig.show()
```

```python
fig=px.parallel_coordinates(data,color=data['class'].astype(int),
                           color_continuous_scale=px.colors.diverging.Tealrose )
fig.show()
```

```python
#seeing insights in the pcp is higly affected by the order of the dimensions that we can see on PCP
#we can change the order of the dimensions though
```

```python
data_sampled=data.sample(n=100)
data_sampled
```

```python
dims=data_sampled.drop(['Country or region','class'],axis=1).columns
fig=px.scatter_matrix(data_sampled,dimensions=dims,color=data_sampled['class'],hover_name=data_sampled['Country or region'])

fig.update_traces(marker=dict(size=4),diagonal_visible=False,showupperhalf=False)#to make the dots smaller
fig.update_layout(width=800,height=800,title='Happiness index SPLOM',font_size=7)
fig.show()
```

```python
#one solution to decrease the amount of overlapping between our data points in our scatter plots is to use sampling
```

```python
#each country is represented by a six dimensional vector
```

```python
#pca is a technique for projection purposes and dimensionality reduction purpose
pca=PCA(2)#we want to shrink the 6 dimensions into 2 dimensions
projected_data=pca.fit_transform(data.drop(['Country or region','class'],axis=1))
```

```python
projected_data
```

```python
fig=px.scatter(projected_data[:,0],projected_data[:,1],color=data['class'],
               labels={'index':'d1','x':'d2'},
               hover_name=data['Country or region'])#everything from the first and second column
fig.show()
```
